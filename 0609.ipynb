```python
import pandas as pd
import json
import datetime
import glob

data = pd.DataFrame()
all_txt_files = glob.glob('/Users/tianyouwu/Desktop/fintech/ParsedData/*.txt')
for file in all_txt_files:
    a = pd.read_json(file)
    data = data.append(a)

b = data[['Name','Time','Text']]
c = b.sort_values('Time')

c

c['Time'] = pd.to_datetime(c['Time']).apply(lambda x:x.strftime('%Y%m%d'))

c

#del c['Time']

#c.query('Time1 == "20200525"')

time_range = c['Time'].drop_duplicates()
data.drop_duplicates('Text',inplace=True)

time_range

for time in time_range:
    df = c.query('Time == @time')
    ans = json.loads(df.to_json(orient='records'))
    with open(f'/Users/tianyouwu/Desktop/fintech/CleanedData/{time}.json','w') as f:
        json.dump(ans,f)
```
